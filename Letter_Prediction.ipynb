## A) Logistic Regression Model

```python
letters_train1["letter"].replace({'B': 1, "not B":0}, inplace=True)
letters_train1
letters_test1["letter"].replace({'B': 1, "not B":0}, inplace=True)
letters_test1
```

```python
import statsmodels.formula.api as smf
logreg = smf.logit(formula = 'letter ~ xbox + ybox + width + height + onpix + xbar + ybar + x2bar + y2bar + xybar + x2ybar + xy2bar + xedge + xedgeycor + yedge + yedgexcor',
                   data = letters_train1).fit()
print(logreg.summary())
```

```python
#make it a classifier with p_threshold = 0.5
from sklearn.metrics import confusion_matrix

y_test_log = letters_test1['letter']
y_prob = logreg.predict(letters_test1.drop(['letter'], axis=1))
y_pred = pd.Series([1 if x > 0.5 else 0 for x in y_prob], index=y_prob.index)

cm5 = confusion_matrix(y_test_log, y_pred)
print ("Confusion Matrix 5 : \n", cm5) 

accuracy_new5 = (cm5[0][0]+cm5[1][1])/np.sum(cm5)
print(accuracy_new5)
FPR5 = cm5[0][1]/(cm5[0][0]+cm5[0][1])
TPR5 = cm5[1][1]/(cm5[1][0]+cm5[1][1])
print(FPR5)
print(TPR5)
```

a). ii). the accuracy of my logistic regression model on the test set = 0.9412

## A) AUC

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test_log, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.title('ROC Curve', fontsize=18)
plt.xlabel('FPR', fontsize=16)
plt.ylabel('TPR', fontsize=16)
plt.xlim([-0.01, 1.00])
plt.ylim([-0.01, 1.01])
plt.plot(fpr, tpr, lw=3, label='Logistic Regression (area = {:0.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=3, linestyle='--')
plt.legend(loc='lower right', fontsize=14)
plt.show()
```

## A) CART model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

y_train_cart = letters_train1['letter']
x_train_cart = letters_train1.drop(['letter'], axis=1)
y_test_cart = letters_test1['letter']
x_test_cart = letters_test1.drop(['letter'], axis=1)

grid_values = {'ccp_alpha': np.linspace(0, 0.02, 50)}
dtc = DecisionTreeClassifier(min_samples_leaf=5, min_samples_split=20, random_state=88)
dtc_cv = GridSearchCV(dtc, param_grid = grid_values, cv=5, verbose =0)
dtc_mod = dtc_cv.fit(x_train_cart,y_train_cart)
```

```python
r2 = dtc_mod.cv_results_['mean_test_score'] 
ccp = dtc_mod.cv_results_['param_ccp_alpha'].data
print(dtc_cv.best_params_)
plt.figure(figsize=(8, 6))
plt.xlabel('ccp alpha', fontsize=16)
plt.ylabel('accuracy', fontsize=16)
plt.scatter(ccp, r2, s=2)
plt.plot(ccp, r2, linewidth=3)
plt.grid(True, which='both')
plt.show()
```

```python
y_pred_cart = dtc_cv.best_estimator_.predict(x_test_cart)
cm_cart = confusion_matrix(y_test_cart, y_pred_cart)
print ("Confusion Matrix : \n", cm_cart) 
accuracy_cart = (cm_cart[0][0]+cm_cart[1][1])/np.sum(cm_cart)
print(accuracy_cart)
```

a). iii). I use 5-fold cross-validation to build the CART model with the specific parameter setting as below.
I select 50 possible cp values in the interval [0,0.02] and let the model to select the best-functional one.
The accuracy of the cart model is 0.9316 and the best cp value is about 0.00082

## bootstrap adoption

```python
from sklearn.metrics import accuracy_score
def bootstrap_validation(test_data, test_label, model, sample=500, random_state=66):
    n_sample = sample
    output_array=np.zeros([n_sample])
    output_array[:]=np.nan
    print(output_array.shape)
    for bs_iter in range(n_sample):
        bs_index = np.random.choice(test_data.index, len(test_data.index), replace=True)
        bs_data = test_data.loc[bs_index]
        bs_label = test_label.loc[bs_index]
        bs_predicted = model.predict(bs_data)
        acc = accuracy_score(bs_label, bs_predicted)
        output_array[bs_iter] = acc
    output_df = pd.DataFrame(output_array)
    return output_df
```

```python
df1 = bootstrap_validation(x_test ,y_test, gbc)
df1 = df1.rename(columns={0: "Boosting"})
df1.describe()
```

```python
df2 = bootstrap_validation(x_test ,y_test, rfm_cv)
df2 = df2.rename(columns={0: "random forest"})
df2.describe()
```

```python
df3 = bootstrap_validation(x_test ,y_test, bag)
df3 = df3.rename(columns={0: "bagging"})
df3.describe()
```

```python
df4 = bootstrap_validation(x_test ,y_test, dtc_new_cv)
df4 = df4.rename(columns={0: "cart"})
df4.describe()
```
